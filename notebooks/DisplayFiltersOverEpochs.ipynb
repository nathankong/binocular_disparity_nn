```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np

import pickle

plot_epoch_idx = [0,10]
```

```python
def plot_filters(filter_pairs, epoch):
    n_filters = filter_pairs.shape[0]
    assert n_filters % 4 == 0
    
    fig, ax = plt.subplots(4, int(n_filters / 4), figsize=(15,6))
    fig.suptitle("Epoch {}".format(epoch), fontsize=14)
    for i in range(4):
        for j in range(int(n_filters/4)):
            filter_pair = filter_pairs[i*7+j]
            concat_filter = np.concatenate((filter_pair[0,:,:], filter_pair[1,:,:]), axis=1)
            
            ax[i,j].imshow(concat_filter, cmap="gray")
            ax[i,j].set_title("Kernel #{}".format(i*(int(n_filters/4))+j+1))

def plot_mid_section(filter_pairs, epoch):
    n_filters = filter_pairs.shape[0]
    assert n_filters % 4 == 0
    
    mid_idx = 9
    fig = plt.figure(figsize=(24,6))
    fig.suptitle("Epoch {}".format(epoch), fontsize=14)
    gs0 = gridspec.GridSpec(4, int(n_filters/4))
    for i in range(4):
        for j in range(int(n_filters/4)):
            filter0 = filter_pairs[i*7+j,0,mid_idx,:]
            filter1 = filter_pairs[i*7+j,1,mid_idx,:]
            
            gs = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec=gs0[i,j])
            ax0 = fig.add_subplot(gs[0])
            ax1 = fig.add_subplot(gs[1])
            ax0.plot(filter0)
            ax1.plot(filter1)
            ax0.set_title("Kernel #{}, Left".format(i*(int(n_filters/4))+j+1), fontsize=10)
            ax1.set_title("Kernel #{}, Right".format(i*(int(n_filters/4))+j+1), fontsize=10)
    fig.tight_layout(rect=[0, 0.03, 1, 0.95])
```

## Visualize Filters

```python
exp_id = "test"
results = "../results/results_{}.pkl".format(exp_id)
results = pickle.load(open(results, "rb"), encoding="latin1")
print(results.keys())
```

```python
epochs_filter = np.array(results["simple_unit_weights"])
print(epochs_filter.shape)
for i in plot_epoch_idx:
    pair = epochs_filter[i]
    plot_filters(pair, i+1)
```
