```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

## Gabor Initialization

```python
k = np.load("../initial_kernels_gabor.npy")
print(k.shape)
```

```python
for i in range(k.shape[0]):
    kernels = k[i]
    fig, (ax1, ax2) = plt.subplots(1,2)
    ax1.imshow(kernels[0,:,:], cmap="gray")
    ax1.set_title("Left Kernel")
    ax2.imshow(kernels[1,:,:], cmap="gray")
    ax2.set_title("Right Kernel")
```
