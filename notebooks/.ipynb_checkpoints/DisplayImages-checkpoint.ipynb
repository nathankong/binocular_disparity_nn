```python
%matplotlib inline
import matplotlib.pyplot as plt

import gzip
import pickle
import numpy as np
```

## Original Pickle File

```python
f = open("/mnt/fs5/nclkong/datasets/bnn_dataset/lytroPatches_30x30.pkl", "rb")
dataset = pickle.load(f, encoding="latin1")
f.close()
dataset_x, dataset_y = dataset
```

```python
print(dataset_x.shape)
print(dataset_y.shape)
```

```python
dataset_x = np.transpose(dataset_x, (0,2,3,1))
```

```python
dataset_x.shape
```

```python
im_idx = 10
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(dataset_x[im_idx,:,:,0], cmap="gray")
ax1.set_title("Left Image")
ax2.imshow(dataset_x[im_idx,:,:,1], cmap="gray")
ax2.set_title("Right Image")
```

## From Image Directory

```python
img_idx = 11
img_fname = "/mnt/fs5/nclkong/datasets/bnn_dataset/crossed/img_{}.npy".format(img_idx)
im = np.load(img_fname)
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(im[:,:,0], cmap="gray")
ax1.set_title("Left Image")
ax2.imshow(im[:,:,1], cmap="gray")
ax2.set_title("Right Image")
```
