```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pickle
```

```python
def plot_loss_and_acc(losses, acc):
    f, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,4))
    
    # Plot losses
    assert len(losses["train"]) == len(losses["validation"])
    train_losses = np.array(losses["train"])
    val_losses = np.array(losses["validation"])
    x_vals = np.arange(len(losses["train"])) + 1
    
    ax1.plot(x_vals, train_losses, label="Train")
    ax1.plot(x_vals, val_losses, label="Validation")
    ax1.set_xlabel("Epoch")
    ax1.set_ylabel("Loss")
    ax1.set_title("Loss Curve")
    ax1.legend()

    # Plot accuracies
    assert len(acc["train"]) == len(acc["validation"])
    train_acc = np.array(acc["train"])
    val_acc = np.array(acc["validation"])
    x_vals = np.arange(len(acc["train"])) + 1
    
    ax2.plot(x_vals, train_acc, label="Train")
    ax2.plot(x_vals, val_acc, label="Validation")
    ax2.set_xlabel("Epoch")
    ax2.set_ylabel("Accuracy")
    ax2.set_title("Accuracy Curve")
    ax2.legend()
```

## Random Initialization

```python
exp_id = "test"
results = "../results/results_{}.pkl".format(exp_id)
results = pickle.load(open(results, "rb"), encoding="latin1")
print(results.keys())
```

```python
plot_loss_and_acc(results["losses"], results["accuracies"])
```

```python
exp_id = "gabor"
results = "../results/results_{}.pkl".format(exp_id)
results = pickle.load(open(results, "rb"), encoding="latin1")
print(results.keys())
```
